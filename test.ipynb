```python
import time

run_time = 5 #52
index = [(i+1) for i in range(0, run_time)]
run_best_fitness_values = []
best_fitness_values = []
avg_fitness_values = []
std_fitness_values = []
solution_list = []


benchmark_function = [
                        #   'F1',
#                           'F2',
#                           'F3',
#                           'F4',
#                              'F5',
#                              'F6',
#                              'F7',
                            #  'F8',
#                             'F9',
#                              'F10',
#                              'F11',
#                              'F12',
#                              'F13',
#                              'F14',
#                              'F15',
#                               'F16',
#                              'F17',
#                              'F18',
#                              'F19',
#                              'F20',
#                              'F21',
#                              'F22',
                             'F23'
                         ]

np.random.seed(42)

SearchAgents = 30
start = time.time()
for j in benchmark_function:
    print(f'run for {j} function')
        
    lowerbound, upperbound, dimension, fitness = fun_info(j)
    Max_iterations = int(np.ceil(100*dimension/SearchAgents)) #10000
    
    for i in range(run_time):
        #print(f'run for {i+1} (st/2nd/rd/th) run time')
        cso_instance = CSO(fitness, P=150, dimension=dimension, plot=True, 
                                                        Tmax=Max_iterations, verbose=True)
        best_solution, best_fitness = cso_instance.execute()
        
        run_best_fitness_values.append(best_fitness)
        
    solution_data = {
        'benchmark_function': j,
        'best_fitness_value': best_fitness,
        'location': best_fitness
    }
    solution_list.append(solution_data)
        
    best_fitness_values.append(best_fitness)
    avg_fitness_values.append(np.mean(run_best_fitness_values))
    std_fitness_values.append(np.std(run_best_fitness_values))
    
    
    print(f'finished for {j} function! with total {Max_iterations} iterations in each {run_time} run time!')
    
    
   
    # Display additional information
    print(f"Function: {j}")
    print("Average Fitness:", np.mean(run_best_fitness_values))
    print("Median Fitness:", np.median(run_best_fitness_values))
    print("Standard Deviation:", np.std(run_best_fitness_values))
    print("Best Score:", best_fitness)

    run_best_fitness_values.clear()

print(f'with final solution {solution_list}')

print("Finish processing in ", time.time() - start,'s')
```
